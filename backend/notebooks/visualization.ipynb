```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# File path for csv File
filepath = "./../data/processed/stacked_df_cleaned.csv"
astro_df = pd.read_csv(filepath, index_col=0, parse_dates=True)
```

```python
astro_df.head()
```

```python
# Plotting the comparison of the team's performance as the home team vs away team
grouping_by_team = astro_df.groupby(by=["team", "is_away"])["ops"].mean()
```

```python
sns.barplot(x="team", y="ops", hue="is_away", data=grouping_by_team.reset_index(), width=0.8)
plt.xticks(fontsize=5.5)
plt.title("Comparison of the team's average performance as the home team vs away team")
plt.show()

# Home games vs Away games (home games has a higher OPS score, but is not really clear...)
```

```python
grouping_by_team = astro_df.groupby(by=["team", "is_away"])["ops_deviation_from_mean"].mean()


sns.barplot(x="team", y="ops_deviation_from_mean", hue="is_away", data=grouping_by_team.reset_index(), width=0.8)
plt.xticks(fontsize=5.5)
plt.title("Amount deviate from the average OPS score")
plt.show()

# Home games vs Away games (home games has a higher OPS score, but is not really clear...)
```
