```python
# imports 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python

pd.set_option('display.max_columns', None)
```

```python
# File path for csv File
filepath = "./../data/processed/stacked_df_cleaned.csv"
astro_df = pd.read_csv(filepath, index_col=0, parse_dates=["game_date"])
```

```python
astro_df.head()
```

```python
# https://baseballsavant.mlb.com/visuals/map?team=&year=2023b
print(astro_df[astro_df['game_date'].dt.year == 2023].groupby("team")["distancetravelled"].sum())
```

```python
train, test = np.split(astro_df, [int(.8*len(astro_df))])
X_train = train[["distancetravelled", "restdays", "is_away", "consecutiveGames", "is_day_game"]]
y_train = train["ops_deviation_from_mean"]
X_test = test[["distancetravelled", "restdays", "is_away", "consecutiveGames", "is_day_game"]]
y_test = test["ops_deviation_from_mean"]
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("MSE: ", mean_squared_error(y_test, y_pred))
print("R2: ", r2_score(y_test, y_pred))

model.coef_
```
